```python
import pandas as pd
import random
```

```python
tickets = pd.read_csv("tickets.csv",sep=",")
```

```python
tickets
```

```python
total_ticket = len(tickets)
total_ticket
```

# Rewards
- 200 X WWV Cowgirl NFT
- 10 x WWV Cowboy NFT
- 0.5 SOL X 10 Winners 
- 0.25 SOL X 20 Winners
- 0.1 SOL X 25 Winners

# Choose 265 ticket number:
From the list:
- first 200 winner ticket will get Cowgirl NFT
- following 10 winner ticket will get Cowboy NFT
- following 10 winner ticket will get 0.5 SOL
- following 20 winner ticket will get 0.25 SOL
- following 25 winner ticket will get 0.1 SOL


```python
all_winners = random.sample(range(total_ticket), 265)
```

```python
all_winners
```

```python
cowgirl_winner_tickets = all_winners[0:200]
cowgirl_winner_tickets
```

```python
textfile = open("cowgirl_winner.txt", "w")
for element in cowgirl_winner_tickets:
    textfile.write(str(element) + "\n")
textfile.close()
```
